# Breast Cancer Wisconsin (Diagnostic) 


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('wdbc.data', header=None)
```

```python
print(dataset.shape)
dataset.head()
```

**Remoção do id**: essa informação não acrescenta em nada para a tarefa de classificação que desejamos


```python
dataset.drop(labels=dataset.columns[0], axis=1, inplace=True)
dataset.head()
```

```python
dataset.isna().sum()
```

 Forma de ver toda a informação no notebook

```python
pd.options.display.max_columns = None
dataset.describe()
```

## Divisão dos dados


```python
X = pd.DataFrame(dataset.iloc[:,1:])
Y = pd.DataFrame(dataset.iloc[:, 0])

encoder_y = LabelEncoder() 
Y = pd.DataFrame( encoder_y.fit_transform(Y))

print(X.shape, Y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
y_train.head()
```

## PCA


```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(5)
```

```python
pca.fit(x_train)
```

```python
pca.explained_variance_ratio_.round(3)
```

```python
x_pca = pd.DataFrame( pca.transform(x_train))
```

```python
x_pca.head()
```

## Feature Scaling
Pré-processamento dos dados


```python

from sklearn.preprocessing import StandardScaler, Normalizer, MaxAbsScaler, MinMaxScaler, RobustScaler
scaler = StandardScaler()# alterar

x_train_norm =  pd.DataFrame(scaler.fit_transform(x_pca)) 
x_train_norm.describe().round(3)
```

## Treinamento

```python
# metricas
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(C=1) 
clf.fit(x_train_norm, y_train)

pred = clf.predict(x_train_norm)
print('accuracy', accuracy_score(y_train,pred))
print('precision', precision_score(y_train,pred ))
print('recall', recall_score(y_train,pred))
print('f1', f1_score(y_train,pred))
confusion_matrix(y_train,pred)
```

## Test

```python
x_test_pca = pca.transform(x_test)
x_test_norm = scaler.transform(x_test_pca)
```

```python
pred = clf.predict(x_test_norm)
print('accuracy', accuracy_score(y_test,pred))
print('precision', precision_score(y_test,pred ))
print('recall', recall_score(y_test,pred))
print('f1', f1_score(y_test,pred))
confusion_matrix(y_test,pred)
```
