# Hello World

Primeiro codigo usando tensorflow

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

Criar rede com apenas uma camada e um nó

```python
model = keras.Sequential(keras.layers.Dense(units=1,input_shape=[1]))
```

Determina o tipo de algoritmo de otimização e o tipo de função de perda que deve ser utilizado no treinamento da rede definida acima

*A seguir, duas formas de definir a mesma coisa*

```python
model.compile(optimizer=keras.optimizers.SGD(), loss=keras.losses.mean_absolute_error)
#model.compile(optimizer='sgd', loss='mean_absolute_error')
```

## Dados

Criando os dados que deverã ser utilizados

```python
x = np.array([-1, 0, 1, 2, 3, 4], dtype=float)
y = np.array([-3, -1, 1, 3, 5, 7], dtype=float)
```

## Treinando a rede

```python
model.fit(x, y, epochs=500)
```

```python
model.predict([10])
```
