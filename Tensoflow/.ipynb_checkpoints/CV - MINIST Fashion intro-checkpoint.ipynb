# MNIST Fashion

Primeiro codigo usando Tensorflow para Visão Computacional.

Neste momento usaremos cada pixel como uma feature (não será convolucional).


```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

##  Carregando o banco

Usando o modulo do Tensorflow para baixar o banco ja dividido em treino e teste

```python
mnist =   tf.keras.datasets.fashion_mnist
(train_images, train_labels) , (test_images, test_labels) = mnist.load_data()
```

```python
plt.imshow(train_images[1])
```

### Normalização das imagens

As imagens serão normalizadas a fim de terem valores entre 0 e 1. Dessa forma estarão prontas para serem utilizadas por um rede neural que se comporta melhor com velores desse tipo

```python
train_images_norm = train_images/255
test_images_norm = test_images/255
```

```python
train_images_norm.max()
```

## Criando o modelo

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten()) # transforma o imagem(matriz) para uma informação unidimensional
model.add(tf.keras.layers.Dense(512, activation=tf.keras.activations.relu))
model.add(tf.keras.layers.Dense(128, activation=tf.keras.activations.relu))
model.add(tf.keras.layers.Dense(10, activation=tf.keras.activations.softmax))
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_images_norm, train_labels,epochs=10)
```

```python
model.evaluate(test_images_norm, test_labels)
```

## Usando callbacks

Utilizando a estratégia de callback durante o treinamento para pará-lo.

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('loss')<0.4):
      print("\nReached 60% accuracy so cancelling training!")
      self.model.stop_training = True

callbacks = myCallback()
mnist = tf.keras.datasets.fashion_mnist
(training_images, training_labels), (test_images, test_labels) = mnist.load_data()
training_images_norm = training_images/255.0
test_images_norm = test_images/255.0
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512, activation=tf.nn.relu),
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')
model.fit(training_images_norm, training_labels, epochs=5, callbacks=[callbacks])
```

**Como pode ser visto, a cada epoch o callback é chamado e assim a condição de parada é avaliada**
